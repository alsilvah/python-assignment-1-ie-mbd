##### First of all I need to import pandas and load Titanic train dataset to a dataframe  

```python
import pandas as pd
```

```python
#Read the dataset correctly using pandas, using an appropriate column as index (1 point)
df=pd.read_csv('train.csv', index_col="PassengerId")
print("Part 1, Q1")
df.head(10)
```

```python
# Compute the percentage of survivors out of total number of passengers and the percentage of men and women out of total number of passengers (0.5 points)

total_passengers=len(df)
total_survivors=len(df[df.Survived==1])
perc_survivors=total_survivors/total_passengers
perc_men=len(df[df.Sex=='male'])/total_passengers
perc_women=len(df[df.Sex=='female'])/total_passengers
print("Answers Part 1 Q2:\n")
print("Percentage of survivors: "+"{:.2%}".format(perc_survivors))
print("Percentage of men of total passengers: "+"{:.2%}".format(perc_men))
print("Percentage of women of total passengers: "+"{:.2%}".format(perc_women))
```

```python
# Compute the percentage of survivors by sex (i.e. the percentage of male passengers that survived and female passengers that survived) and the sex distribution of survivors (i.e. percentage of survivors that were men and percentage of survivors that were women) (0.5 points)
perc_men_surv=(len(df[(df.Sex=='male')&(df.Survived==1)]))/(len(df[df.Sex=='male']))
perc_women_surv=len(df[(df.Sex=='female')&(df.Survived==1)])/(len(df[df.Sex=='female']))
print ("Percentage of survivors in men: "+"{:.2%}".format(perc_men_surv))
print ("Percentage of survivors in women: "+"{:.2%}".format(perc_women_surv))
```

```python
# Compute the sex distribution of survivors (i.e. percentage of survivors that were men and percentage of survivors that were women) (0.5 points)¶
perc_surv_men=len(df[(df.Sex=='male')&(df.Survived==1)])/len(df[df.Survived==1])
perc_surv_women=len(df[(df.Sex=='female')&(df.Survived==1)])/len(df[df.Survived==1])
print ("Percentage of men in survivors: "+"{:.2%}".format(perc_surv_men))
print ("Percentage of women in survivors: "+"{:.2%}".format(perc_surv_women))
```

```python
## Q4: Display in a 2 x 2 DataFrame the probability of being male/female and surviving/not surviving (0.5 points)
df_dist_2=df[["Sex","Survived"]].pivot_table(index="Sex",columns="Survived",aggfunc=len).div(len(df.index)).applymap(lambda x: "{:.2f}%".format(100*x))
```

```python
df_dist_2.columns=["Not Survivor","Survivor"]
```

```python
print("Answer Part 1 Q4:\n")

df_dist_2
```

```python
# I created a Pivot table adding Pclass in order to have the details of survivors grouped by Sexa and Pclass
df_dist_3=df[["Sex","Survived","Pclass"]].pivot_table(index=["Sex","Pclass"],columns="Survived",aggfunc=len).div(len(df.index)).applymap(lambda x: "{:.2f}%".format(100*x))
df_dist_3.columns=["Not Survivor","Survivor"]
```

```python
print("Answer Part 1 Q5:\n")
df_dist_3
```

#### To complete the age ranges, my strategy will be to first obtain the average age by gender and key, and impute the null values for these fields, that is, if there is a null value given a class and a specific sex, I will change that null value for the mean calculated for that record with that sex and that class (considering the Sex and Pclass columns as sex and class)

```python
df.groupby(["Pclass","Sex","Parch"])["Age"].agg(["mean","std"])
```

```python
df
```

```python
# I created a dataframe with average age of  the entire dataset grouped by Sex,Pclass and Parch
# I did the conversion  to int64 and then to float64 so that it is in "XX.00" format just like in the original DF

avg_ages_per_sex_class_df=(pd.DataFrame(df.groupby(["Sex","Pclass","Parch"]).agg('mean')["Age"])).astype('int64').astype('float64')
```

```python
#avg_ages_per_sex_class_df=pd.DataFrame(avg_ages_per_sex_class)
avg_ages_per_sex_class_df
```

```python
# Before the merge operation of the two dataframes, I decided to reset the index of both dataframes
df=df.reset_index()
avg_ages_per_sex_class_df=avg_ages_per_sex_class_df.reset_index()
```

#### Finally I will merge the DataFrame df with avg_ages_per_sex_class_df by Sex, Pclass and Parch (applying left join)
#### I will drop the new column with averages and rename column Age_x to its original column name Age. And finally I checked every field counting its null values

```python
df = (df.merge(avg_ages_per_sex_class_df, on = ['Sex','Pclass','Parch'], how='left'))
df['Age_x'] = df.Age_x.fillna(df.pop('Age_y'))
df=df.rename(columns={'Age_x':'Age'})
df=df.set_index("PassengerId")
```
